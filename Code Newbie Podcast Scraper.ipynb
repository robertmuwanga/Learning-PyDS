```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup

# Define CodeNewbie RSS feed
url = 'http://feeds.codenewbie.org/cnpodcast.xml'
```

```python
# Download xml page and load it into beautiful soup
response = requests.get(url)
codenewbie = BeautifulSoup(response.text, features='lxml')
```

```python
# Let's build a dataframe with basic information on the podcasts
podcast_items = codenewbie.findAll('item')
podcast_df = pd.DataFrame(index = range(0, len(podcast_items)),
                         columns = ['date', 'title', 'duration', 'size', 'url'])

for index, item in enumerate(podcast_items) :
    podcast_df.loc[index, ] = [
        re.search(re.compile('\d{2}.[A-Za-z]{3}.\d{4}'), item.pubdate.text).group(), 
        item.title.text, 
        item.find('itunes:duration').text, 
        item.enclosure['length'], 
        item.enclosure['url']]

podcast_df = podcast_df.set_index('date')
```
